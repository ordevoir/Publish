Напишем функцию, которая возвращает дискретный вариант логнормального распределения:

```python
def get_lognormal_dist(mu=0.5, sigma=0.5, start=0.1, stop=10., num=1000):
    xs = np.linspace(start, stop, num)
    numerator = np.exp(-((np.log(xs) - mu)**2) / (2 * sigma**2))
    denominator = (xs * sigma * np.sqrt(2 * np.pi))
    probabilities = numerator / denominator
    probabilities /= probabilities.sum()
    return {"xs": xs, "ps": probabilities}
```

```python
mean_1, mean_2, mean_3 = 0., 1., 2.,
sigma_1, sigma_2, sigma_3 = 0.5, 1., 2.

lnDist_1 =get_lognormal_dist(mean_1, sigma_1)
lnDist_2 =get_lognormal_dist(mean_2, sigma_1)
lnDist_3 =get_lognormal_dist(mean_2, sigma_3)
lnDist_4 =get_lognormal_dist(mean_3, sigma_2)


print(lnDist_1["ps"].sum())
```

```python

plt.plot(lnDist_1["xs"], lnDist_1["ps"], label=f"μ={mean_1}, σ={sigma_1}")
plt.plot(lnDist_2["xs"], lnDist_2["ps"], label=f"μ={mean_2}, σ={sigma_1}")
plt.plot(lnDist_3["xs"], lnDist_3["ps"], label=f"μ={mean_2}, σ={sigma_3}")
plt.plot(lnDist_4["xs"], lnDist_4["ps"], label=f"μ={mean_3}, σ={sigma_2}")
plt.title("log-normal distribution")
plt.legend(loc="upper right")
plt.xlabel("x")
plt.ylabel("PDF")
plt.show()
```
