```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ordevoir-dark")
```

```python
def get_normal_dist(mu=0., sigma=1., start=-4., stop=4., num=100):
    xs = np.linspace(start, stop, num)
    numerator = np.exp((-(xs - mu)**2 / (2 * sigma**2)))
    denominator = sigma * np.sqrt(2 * np.pi)
    return {"xs": xs, "ps": numerator / denominator}
```

```python
mu_1, mu_2 = 0., 2.
sigma_1, sigma_2, sigma_3 = 0.5, 1., 2

normDist_1 = get_normal_dist(mu_1, sigma_1)
normDist_2 = get_normal_dist(mu_1, sigma_2)     # Стандартное нормальное распределение
normDist_3 = get_normal_dist(mu_1, sigma_3)
normDist_4 = get_normal_dist(mu_2, sigma_2)

plt.plot(normDist_1["xs"], normDist_1["ps"], label=f"μ={mu_1}, σ={sigma_1}")
plt.plot(normDist_2["xs"], normDist_2["ps"], label=f"μ={mu_1}, σ={sigma_2}")
plt.plot(normDist_3["xs"], normDist_3["ps"], label=f"μ={mu_1}, σ={sigma_3}")
plt.plot(normDist_4["xs"], normDist_4["ps"], label=f"μ={mu_2}, σ={sigma_1}")

plt.title("Normal distribution")
plt.xlabel("x")
plt.ylabel("PDF")
plt.legend(loc="upper right")
plt.grid()
plt.show()
```
