```python
import math
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ordevoir-dark")
```

Напишем функцию, вычисляющую число сочетаний из `n` по `k`:

```python
def combinations(n, k):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n-k))
```

Формула Бернулли:

```python
def binomial(n, p, k):
    return combinations(n, k) * (p**k) * ((1 - p)**(n - k))
```

```python
def B(p, n):
    probabilities = np.empty(n+1, dtype=np.float32)
    for k in range(n+1):
        probabilities[k] = binomial(n, p, k)
    distribution = {
        "ks": np.arange(n+1),
        "ps": probabilities,
    }
    return distribution
```

```python
p1, p2, p3, p4 = 0.2, 0.5, 0.8, 0.95
binDist_1 = B(p=p1, n=100)
binDist_2 = B(p=p2, n=100)
binDist_3 = B(p=p3, n=100)
binDist_4 = B(p=p4, n=100)

plt.plot(binDist_1["ks"], binDist_1["ps"], label=f"$p_1=${p1}")
plt.plot(binDist_2["ks"], binDist_2["ps"], label=f"$p_2=${p2}")
plt.plot(binDist_3["ks"], binDist_3["ps"], label=f"$p_3=${p3}")
plt.plot(binDist_4["ks"], binDist_4["ps"], label=f"$p_4=${p4}")
plt.title('Binomial distribution')
plt.xlabel('$k$')
plt.ylabel("$P(X = k)$")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```
