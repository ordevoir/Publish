```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ordevoir-dark")
```

```python
def G(p, n):
    probabilities = np.empty(n+1, dtype=np.float32)
    for k in range(n+1):
        probabilities[k] = p * (1 -p) ** (k)
    distribution = {
        "ks": np.arange(n+1),
        "ps": probabilities,
    }
    return distribution
```

```python
p1, p2, p3, p4 = 0.01, 0.05, 0.2, 0.5
binDist_1 = G(p=p1, n=10)
binDist_2 = G(p=p2, n=10)
binDist_3 = G(p=p3, n=10)
binDist_4 = G(p=p4, n=10)

plt.plot(binDist_1["ks"], binDist_1["ps"], label=f"$p_1=${p1}")
plt.plot(binDist_2["ks"], binDist_2["ps"], label=f"$p_2=${p2}")
plt.plot(binDist_3["ks"], binDist_3["ps"], label=f"$p_3=${p3}")
plt.plot(binDist_4["ks"], binDist_4["ps"], label=f"$p_4=${p4}")
plt.title('Binomial distribution')
plt.xlabel('$k$')
plt.ylabel("$P(X = k)$")
plt.legend(loc="upper right")
plt.grid()
plt.show()
```
